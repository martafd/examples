```python
import pandas as pd
pd.set_option('display.max_columns', 100)

data = pd.read_csv('../../../Downloads/train.csv')
data.head()
```

- So, as **data_description.txt** says some of the numerical data represent categorical variables (for example MSSubClass)



- That is why, we can assume that there just 2 types of data - numerical and categorical. 
- Let's build the working flow to check the data quality.

---
Let's consider feature **LotArea** as example for numerical features:
we can check the IQR, mean-median metrics and check the distribution to understand if new data is in the same range.

```python
print('Main statistical metrics for feature LotArea')
pd.DataFrame(data.LotArea.describe()).rename(columns={'LotArea': 'LotArea_old_data'}).join(
    pd.DataFrame(new_data.LotArea.describe()).rename(columns={'LotArea': 'LotArea_new_data'}))
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(figsize=(20, 5))
plt.hist([data.LotArea, new_data.LotArea], bins=50, label=['old data', 'new data'])
plt.legend(loc='upper right')
plt.ylabel('amount')
plt.xlabel('lot size in square feet')
plt.title('Distribution of Lot size')
plt.show()
```

- Additionaly, to automate checking, we can apply statistical t-test for numerical values using **stats** package of **scipy** library.

```python
# check unique values of LotArea
data.LotShape.describe()
```
